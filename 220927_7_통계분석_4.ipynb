```python
# SOM : 자율 학습에 의한 클러스터링 수행
# Map 설정 > fit  > 클러스터링
```

```python
import pandas as pd
import numpy as np
iris = pd.read_csv('data/iris.csv')
iris
```

```python
y = iris['target']
X = iris.drop('target', axis = 1)
```

```python
# SOM은 이상치에 민감하기 때문에, Feature Scaling or StandardScaler 수행 필요
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,1))
X = scaler.fit_transform(X)
```

```python
X
```

```python
from sklearn_som.som import SOM
```

```python
iris_som = SOM(m=3, n=1, dim=4) # 3 * 1, dim(변수의 개수)
iris_som.fit(X)
```

```python
predictions = iris_som.predict(X)
predictions
```

```python
iris['cluster'] = predictions
iris
```

```python
# 원본 target으로 시각화
# 군집결과 시각화
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(iris,
             diag_kind = 'kde',
             hue = 'cluster',
             palette = 'bright')
plt.show()
```

```python
# 17회 기출
# 데이터를 7:3으로 나누어라
# train 데이터셋을 som 알고리즘으로 군집화하고 test데이터에 대한 오분류표를 구하라
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify = y, test_size = 0.3, random_state = 2021)
```

```python
som = SOM(m=3, n=1, dim=4)
```

```python
som.fit(X_train)
```

```python
y_pred = som.predict(X_test)
```

```python
result = pd.DataFrame()
result['y_test'] = y_test
result['y_pred'] = y_pred
result
```

```python
# 범주형 값을 숫자로 변환 필요
```

```python
result.loc[result['y_test'] == 'Iris-setosa', 'target_num'] = 2
result.loc[result['y_test'] == 'Iris-virginica', 'target_num'] = 0
result.loc[result['y_test'] == 'Iris-versicolor', 'target_num'] = 1
```

```python
result
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(result['target_num'], result['y_pred'])
```

```python
confusion_matrix(result['target_num'], result['y_pred'])
```
