# 군집분석 과제 

```python
import pandas as pd
```

```python
df = pd.read_csv("data/USArrests.csv")
df
```

위의 데이터에서 Murder	Assault	UrbanPop	Rape를 가지고 아래와 같이  K-means 군집분석을 진행하세요 
- 1) 최적의 k 값 찾으세요. 
- 2) 콜린스키하라바츠스코어 구하세요. 
- 3) SSE구하세요 
- 4) 군집분석한 결과를 해석하세요 

```python
from sklearn.cluster import KMeans
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, 
             diag_kind = 'kde',
             hue ='Unnamed: 0',
             palette = 'bright')
plt.show()
# 아래 그래프를 보니, 범주가 많아 의미가 없다.
```

```python
X = df.drop('Unnamed: 0', axis = 1)
X
```

```python
# K-Means 군집화 실행
km = KMeans(n_clusters = 2, random_state = 1)
km.fit(X)
```

```python
km.labels_
```

```python
# 군집분석 평가 척도 : 콜린스키하라바츠스코어
from sklearn.metrics import calinski_harabasz_score
calinski_harabasz_score(X, km.labels_)
```

```python
for k in range(2,11):
    km = KMeans(n_clusters = k, random_state = 1).fit(X)
    score = calinski_harabasz_score(X, km.labels_)
    print("K : ", k, " ", score)
```

```python
# SSE 구하세요
import matplotlib.pyplot as plt
def elbow(X):
    sse=[]
    for i in range(1, 11):
        km=KMeans(n_clusters=i, random_state=1)
        km.fit(X)
        sse.append(km.inertia_)
    
    plt.plot(range(1,11), sse, marker='o')
    plt.xlabel('The Number of Clusters')
    plt.ylabel('SSE')
    plt.show()
    print(sse)
```

```python
km.inertia_
```

```python
elbow(X)
```

```python
# 군집분석한 결과를 해석하세요
# 콜린스키하라바츠스코어 : 최적의 군집의 수는 9개로 보이나
# 엘보우 방법으로 : 3~4로 갈때 기울기가 소멸
# 최적의 군집의 수는 2~3
```

```python
# SSE
km = KMeans(n_clusters = 3, random_state=1).fit(df.iloc[:,1::])
print("SSE:{}".format(km.inertia_))
```

```python
df['cluster'] = km.labels_
```

```python
df
```

```python
sns.pairplot(df, 
             diag_kind = 'kde',
              hue = 'cluster',
               palette = 'bright')
```

```python
df.groupby('cluster').mean()
```
