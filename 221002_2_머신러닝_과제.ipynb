# 11.Ridge, Lasso, Elastic 선형회귀분석 과제 

변수 설명 

id - 판매된 각 주택의 고유 ID <br>
date  - 주택 판매 날짜<br>
price - 판매된 각 주택의 가격<br>
bedrooms  - 침실 수<br>
bathrooms  - 화장실이 있지만 샤워 시설은 없는 방을 0.5로 설명하는 욕실 수<br>
sqft_living - 아파트 내부 생활 공간의 평방 피트<br>
sqft_lot - 토지 공간의 평방 피트수<br>
floors  - 층 수<br>
waterfront - 아파트가 해안가를 내려다보고 있는지 여부에 대한 더미 변수 - 범주형<br>
view - 부동산에 대한 뷰가 얼마나 좋았는지에 대한 0에서 4 사이의 인덱스 - 범주형<br>
condition - 아파트 상태에 대한 1에서 5까지의 지수, - 범주형 <br>
grade - 1-13까지의 수로,1-3은 건축 및 설계 미달, 7은 시공 및 설계 평균, 11-13은 시공 및 설계 수준 높은 수준이다. - 범주형<br>
sqft_above - 지면보다 높은 내부 주택 공간의 평방 피트수<br>
sqft_basement - 지면 아래에 있는 내부 주택 공간의 평방 피트수<br>
yr_built - 집이 처음 지어진 연도<br>
yr_renovated - 집의 마지막 리모델링 연도<br>
sqft_living15 - 가장 가까운 이웃 15명을 위한 내부 주택 생활 공간의 평방 피트<br>
sqft_lot15 - 가장 가까운 15개 이웃 토지 부지의 평방 피트수<br>

```python
import pandas as pd 
df = pd.read_csv("data/kc_house_data.csv")
df
```

## 전처리 :  id, date 삭제 

```python
df = df.drop(['id', 'date'], axis = 1)
```

## 범주형 변수 전처리 

```python
df.info()
df.waterfront.value_counts()
df.view.value_counts()
df.condition.value_counts()
df.grade.value_counts()
df = pd.get_dummies(df, columns = ['waterfront', 'view', 'condition','grade'])
```

```python
df
```

## 종속변수 y = price , 나머지는 독립변수 X로 설정 

```python
X = df.drop(['price'], axis = 1)
y = df['price']
```

```python
y
```

## 홀드아웃 기법을 사용하여 데이터 분할 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 0)
```

## Ridge 회귀분석을 하여 R2 값과 RMSE값을 구하고 설명하세요. 

```python
import warnings
warnings.filterwarnings(action = 'ignore')

from sklearn.linear_model import Ridge

ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
import numpy as np
print("사용한 데이터 컬럼 수 : ", np.sum(ridge.coef_!=0))
print("훈련 데이터 셋 성능 : ", ridge.score(X_train, y_train))
print("테스트 데이터 셋 성능 : ", ridge.score(X_test, y_test))
```

```python
# alpha 1로 하였을 경우, R2는 0.675임을 알 수 있다.
```

```python
from sklearn.model_selection import GridSearchCV

ridge = Ridge()
param_grid = {'alpha':(0.0001,0.0005,0.001,0.005, 0.01,0.05,0.1, 0.15, 0.2, 0.25, 0.3,0.5,1,2,3)}
model = GridSearchCV(ridge, param_grid)
model.fit(X_train, y_train)
```

```python
model.best_estimator_ # 이전과 동일하다.
```

```python
y_pred = model.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(mse)
print(rmse)
```

```python
# rmse 210,992으로 확인된다. 이는 예측값과 실제값과 210,992만큼 차이가 날 수 있다.
```

## 스케일링 한 데이터셋의 결과와 스케일링 하지 않은 데이터 셋의 결과를 비교해보세요.
> Tip : X_train 데이터를 fit_transform 하여 전처리, X_test는 transform 사용하여 전처리 <br>
>  X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns,index=X_train.index)

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

train_X_scaled = scaler.fit_transform(X_train)
```

```python
train_X_scaled
```

```python
test_X_scaled = scaler.transform(X_test)
```

```python
model = Ridge()
model.fit(train_X_scaled, y_train)
```

```python
## R2 
print("사용한 데이터 컬럼 수 :", np.sum(model.coef_!=0))
print("훈련 데이터 셋 성능 : ", model.score(train_X_scaled, y_train))
print("테스트 데이터 셋 성능 : ", model.score(test_X_scaled, y_test))
```

```python
y_pred = model.predict(test_X_scaled)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(mse)
print(rmse)
```

```python
# rmse 210,475으로 확인된다. 이는 예측값과 실제값과 210,475만큼 차이가 날 수 있다.
```

## 회귀 계수 중 집값과 상관성이 높은 3가지 변수를 해석하세요 

```python
coef = pd.DataFrame(model.coef_, index = X.columns, columns = ['coef'])
coef
```

```python
coef['abs_coef'] = abs(coef.coef)
coef.sort_values('abs_coef', ascending = False)
```

```python
# 가장 상관관계가 높은 변수는 yr_built, grade_11, grade_10 이다. 그러나  grade 10~12는 동일한 범주로, sqft_living 까지 
# 영향을 미치는 요소로 판단한다.
```
