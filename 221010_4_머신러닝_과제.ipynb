```python
import pandas as pd 
from sklearn.datasets import load_boston
boston= load_boston()
X = pd.DataFrame(boston.data,columns=boston.feature_names)
y = boston.target
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2022)
```

```python
# 배깅, 랜덤, 부스팅(회귀)
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, AdaBoostRegressor
model_bag_reg = BaggingRegressor() # n_estimators = 10
model_bag_reg.fit(X_train, y_train)

model_rad_reg = RandomForestRegressor()
model_rad_reg.fit(X_train, y_train)

model_boo_reg = AdaBoostRegressor()
model_boo_reg.fit(X_train, y_train)



print('배깅_train : ' , model_bag_reg.score(X_train, y_train))
print('배깅_test : ' , model_bag_reg.score(X_test, y_test))
print('')
print('랜덤_train : ' , model_rad_reg.score(X_train, y_train))
print('랜덤_test : ' , model_rad_reg.score(X_test, y_test))
print('')
print('부스팅_train : ' , model_boo_reg.score(X_train, y_train))
print('부스팅_test : ' , model_boo_reg.score(X_test, y_test))
```

```python
df_t = pd.read_csv("data/titanic.csv")
df_t
```

```python
df_t.info()
```

```python
# 전처리
df_t['age'] = df_t['age'].fillna(df_t['age'].median())
df_t['embarked'] = df_t['embarked'].fillna(df_t['embarked'].value_counts().index[0])
df_t = df_t.drop(['class', 'embark_town'], axis = 1)
df_t = pd.get_dummies(df_t, ['sex','embarked'])
```

```python
df_t
```

```python
df_t.info()
```

```python
X = df_t.drop(['survived'], axis = 1)
y = df_t['survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 2022)
```

```python
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, AdaBoostClassifier
model_bag_cls = BaggingClassifier() # n_estimators = 10
model_bag_cls.fit(X_train, y_train)

model_rad_cls = RandomForestClassifier()
model_rad_cls.fit(X_train, y_train)

model_boo_cls = AdaBoostClassifier()
model_boo_cls.fit(X_train, y_train)



print('배깅_train : ' , model_bag_cls.score(X_train, y_train))
print('배깅_test : ' , model_bag_cls.score(X_test, y_test))
print('')
print('랜덤_train : ' , model_rad_cls.score(X_train, y_train))
print('랜덤_test : ' , model_rad_cls.score(X_test, y_test))
print('')
print('부스팅_train : ' , model_boo_cls.score(X_train, y_train))
print('부스팅_test : ' , model_boo_cls.score(X_test, y_test))
```

```python
print(model_bag_cls.get_params())
print('')
print(model_rad_cls.get_params())
print('')
print(model_boo_cls.get_params())
```

```python
# GridSearch
from sklearn.model_selection import GridSearchCV
import numpy as np

param_bag = {'n_estimators' : range(10,110,10)}
param_rad = {'max_depth':range(2,11), 'min_samples_leaf':range(1,11)}
param_boo = {'learning_rate': np.arange(0.1,1.1,0.1)}
grid_bag_cls = GridSearchCV(BaggingClassifier(), param_grid=param_bag, n_jobs=1)
grid_rad_cls = GridSearchCV(RandomForestClassifier(), param_grid=param_rad, n_jobs=1)
grid_boo_cls = GridSearchCV(AdaBoostClassifier(), param_grid=param_boo, n_jobs=1)
```

```python
# fit
grid_bag_cls.fit(X_train,y_train)
grid_rad_cls.fit(X_train,y_train)
grid_boo_cls.fit(X_train,y_train)
```

```python
print(grid_bag_cls.best_estimator_)
print(grid_rad_cls.best_estimator_)
print(grid_boo_cls.best_estimator_)
```

```python
# Score
print('배깅의 R^2 : ', grid_bag_cls.score(X_test, y_test))
print('랜덤포레스트의 R^2 : ', grid_rad_cls.score(X_test, y_test))
print('부스팅의 R^2 : ', grid_boo_cls.score(X_test, y_test))
```

```python
# ROC_AUC
from sklearn.metrics import plot_roc_curve, roc_auc_score

plot_roc_curve(grid_bag_cls.best_estimator_, X_test, y_test)
plt.plot([0,1], [0,1])

plot_roc_curve(grid_rad_cls.best_estimator_, X_test, y_test)
plt.plot([0,1], [0,1])

plot_roc_curve(grid_boo_cls.best_estimator_, X_test, y_test)
plt.plot([0,1], [0,1])

plt.show()
```

```python
print('roc_auc_score of bagging :', roc_auc_score(y_test, grid_bag_cls.best_estimator_.predict_proba(X_test)[:,1]))
print('roc_auc_score of randomforest :', roc_auc_score(y_test, grid_rad_cls.best_estimator_.predict_proba(X_test)[:,1]))
print('roc_auc_score of boosting :', roc_auc_score(y_test, grid_boo_cls.best_estimator_.predict_proba(X_test)[:,1]))
```

```python
# 시험자료에 대한 ROC_AUC 점수를 보았을 때 랜덤포레스트가 가장 좋았고 이 때의 값은 0.866이다.
```
